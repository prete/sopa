### Herramientas de Python3 para el procesamiento de reportes de SoPI

### Carga de librerias

La herramienta requiere las siguientes librerías.

- [matplotlib](http://matplotlib.org/) para generación de gráficos.
- [numpy](http://www.numpy.org/) para el procesamiento numérico y matricial.
- [lxml](http://lxml.de/) para el procesamiento de documentos XML y HTML

```python
import sopa

# Valores de longitud de onda por banda para Landsat8
lambdas = [482, 561, 655, 864, 1608, 2200]

with open('reporte-areas-entrenamiento.htm') as archivo:
    contenido = archivo.read()
    reportes = sopa.procesar(contenido)
    reportes.plot_medias(lambdas, "Longitud de londa [nm]", "Relectancia [Arb.]", "Firmas espectrales")
```
